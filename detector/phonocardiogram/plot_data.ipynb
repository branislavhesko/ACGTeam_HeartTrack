```python
from scipy.io import loadmat
import matplotlib.pyplot as plt


pcg = loadmat('C:/Users/vojta/Downloads/PCG_dataset.mat')['PCG_dataset']
```

```python
signal = pcg[0, 0][0]
peak_locs = pcg[0, 0][1]

plt.plot(signal[:20000])
plt.scatter(peak_locs[:6], signal[peak_locs[:6]], c='red')
```

```python
from scipy.signal import butter, filtfilt

fs = 3000 
low_cutoff = 30 
high_cutoff = 70  


def bandpass_filter(data, low_cutoff, high_cutoff, fs, order=2):
    nyquist = 0.5 * fs
    low = low_cutoff / nyquist
    high = high_cutoff / nyquist
    b, a = butter(order, [low, high], btype="band")
    
    print(b, a)
    print(data.shape)
    
    y = filtfilt(b, a, data)
    return y

filtered_signal = bandpass_filter(signal.flatten(), low_cutoff, high_cutoff, fs)
plt.plot(filtered_signal[:20000])

for loc in peak_locs[:6]: 
    plt.axvline(x=loc, color="red", linestyle="--", label="Peak" if loc == peak_locs[0] else "")
```
